```python
# Turn text to lowercase
def lowerFun(text):
    return text.lower()

# Remove punctuations
def removePunctuation(text):
    text.replace("\n","")
    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    for c in text:
        if c in punc:
            text = text.replace(c, "")
    return text
    
def removePunctuationList(text_list):
    for i in range(len(text_list)):
        text_list[i] = removePunctuation(text_list[i])
    return text_list

def processTexts(lang='en'):
    df[lang] = df[lang].apply(lowerFun)
    # Split the data into sentences: 
    df[lang] = df[lang].apply(tokenize.sent_tokenize) 
    df[lang] = df[lang].apply(removePunctuationList)
```

```python
# 3. creez exemple de clasificare cu cate 500 de cuvinte fiecare, fara sa rupem propozițiile
```

```python
def create_classification_examples(df, example_size=200, lang='en'):
    # returns list of classification examples
    result = []
    current_example = ""
    for speech_list in df[lang]:
        for sentence in speech_list:
            current_example += sentence
            if len(current_example) > example_size:
                result.append(current_example)
                current_example = ""
    if current_example != "":
        result.append(current_example)
    return result
```

```python
eng_class = create_classification_examples(eng_df, 250)
irs_class = create_classification_examples(irs_df, 250)
wal_class = create_classification_examples(wal_df, 250)
```

```python
# 4. o prima clasificare tf-idf cu vocabulary=english_stop_words
```

```python
class_df = pd.read_csv('../dataset/classification_df.csv') 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, matthews_corrcoef, balanced_accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from sklearn.neural_network import MLPClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    class_df['text'], class_df['label'], test_size=0.2, random_state=42)
```

```python
def predict(vectorizer, X_train, X_test, y_train, classifier=LogisticRegression(max_iter=1000)):
    X_train_tfidf = vectorizer.fit_transform(X_train)
    X_test_tfidf = vectorizer.transform(X_test)

    classifier.fit(X_train_tfidf, y_train)
    
    # Predict on the test set
    y_pred = classifier.predict(X_test_tfidf)
    return y_pred
```

```python
def plot_confusion_matrix(conf_matrix, lang='en'):
    labels = ['eng', 'irs', 'wal']
    plt.figure(figsize=(4, 2))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix '+lang)
    plt.show()
```

```python
vectorizer = TfidfVectorizer(vocabulary=list(ENGLISH_STOP_WORDS))

y_pred = predict(vectorizer, X_train, X_test, y_train)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
plot_confusion_matrix(conf_matrix)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

b_accuracy = balanced_accuracy_score(y_test, y_pred)
print(f"Balanced Accuracy: {b_accuracy:.4f}")

mcc = matthews_corrcoef(y_test, y_pred)
print(f"Matthews Correlation Coefficient (MCC): {mcc:.4f}")
```

```python
vectorizer = TfidfVectorizer(stop_words='english')

y_pred = predict(vectorizer, X_train, X_test, y_train)
```

```python
# 5. folosim shap https://github.com/shap/shap pentru a scoate niste grafice de explainability
```

```python
import shap
```

```python
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)

# Train Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

# Train MLPClassifier
mlp_model = MLPClassifier()
mlp_model.fit(X_train, y_train)

# Explain for Logistic Regression
explainer_lr = shap.LinearExplainer(lr_model, X_train, feature_dependence="independent")
shap_values_lr = explainer_lr.shap_values(X_test)

# Explain for MLPClassifier
explainer_mlp = shap.KernelExplainer(mlp_model.predict_proba, X_train[:100])  # Use a subset for KernelExplainer
shap_values_mlp = explainer_mlp.shap_values(X_test[:10])  # Use a subset to save time

# Get feature names
feature_names = vectorizer.get_feature_names_out()
```

```python
shap.summary_plot(shap_values_lr, X_test, feature_names=feature_names)
```

```python
shap.summary_plot(shap_values_mlp, X_test[:10], feature_names=feature_names)
```

```python
shap.initjs()
```

```python
shap.force_plot(explainer_lr.expected_value[0], shap_values_lr[0][0], feature_names=feature_names)
```

```python
shap.force_plot(explainer_mlp.expected_value[0], shap_values_mlp[0][0], feature_names=feature_names)
```

```python
languages = ['da', 'de', 'en', 'es', 'fr', 'it', 'nl']

for l in languages:
    processTexts(l)

eng_df = df[df['state']=='England'].sample(n=250)
irs_df = df[df['state']=='Ireland'].sample(n=250)
wal_df = df[df['state']=='Wales'].sample(n=193)

# Write preprocessed data to file

eng_df.to_csv('../dataset/eng_df_preprocessed.csv')
irs_df.to_csv('../dataset/irs_df_preprocessed.csv')
wal_df.to_csv('../dataset/wal_df_preprocessed.csv')
```

```python
for l in languages:
    eng_class = create_classification_examples(eng_df, 250, l)
    irs_class = create_classification_examples(irs_df, 250, l)
    wal_class = create_classification_examples(wal_df, 250, l)
    
    label = len(eng_class) * ['eng'] + len(irs_class) * ['irs'] + len(wal_class) * ['wal']
    texts = eng_class + irs_class + wal_class
    pd.DataFrame({'text' : texts, 'label' : label}).sample(frac=1).to_csv('../dataset/CLASS_DFS/' + l + '_classification_df.csv')
```

```python
vectorizer = TfidfVectorizer()

for l in languages:
    class_df = pd.read_csv('../dataset/CLASS_DFS/' + l + '_classification_df.csv')
    X_train, X_test, y_train, y_test = train_test_split(
            class_df['text'], class_df['label'], test_size=0.2, random_state=42)
    y_pred = predict(vectorizer, X_train, X_test, y_train)
    conf_matrix = confusion_matrix(y_test, y_pred)
    plot_confusion_matrix(conf_matrix, l)
    
    b_accuracy = balanced_accuracy_score(y_test, y_pred)
    print(f"Balanced Accuracy for {l}: {b_accuracy:.4f}")
    
    mcc = matthews_corrcoef(y_test, y_pred)
    print(f"Matthews Correlation Coefficient (MCC) for {l}: {mcc:.4f}")

    
```
